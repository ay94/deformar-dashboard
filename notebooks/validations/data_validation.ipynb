```python
def custom_summary(df, col):
    summary = {}
    
    counts = df[col].value_counts()
    percents = df[col].value_counts(normalize=True) * 100
    summary[col] = pd.DataFrame({'Count': counts, 'Percent': percents.round(2)})
    return summary

summary = custom_summary(analysis_data, [
    'Strict True Entities', 'Strict Pred Entities',
    'True Entities', 'Pred Entities'
])
summary['Strict True Entities']  # or loop through all
```
