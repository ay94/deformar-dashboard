```python
import plotly.express as px
import pandas as pd
import numpy as np

# Generate sample data (replace with actual values)
np.random.seed(42)
num_points = 500
variability = np.random.uniform(0, 0.5, num_points)
confidence = np.random.uniform(0, 1, num_points)
correctness = np.random.uniform(0, 1, num_points)  # Simulating correctness scores

# Create DataFrame
df = pd.DataFrame({
    "Variability": variability,
    "Confidence": confidence,
    "Correctness": correctness
})

# Scatter plot with color and size based on correctness
fig = px.scatter(df, x="Variability", y="Confidence", 
                 color="Correctness", 
                 size="Correctness", 
                 color_continuous_scale="blues", 
                 title="Confidence vs Variability (Dataset Cartography)")

# Add annotations for Easy-to-Learn, Ambiguous, and Hard-to-Learn regions
fig.add_annotation(x=0.2, y=0.9, text="Easy-to-Learn", showarrow=False, font=dict(size=14, color="black"))
fig.add_annotation(x=0.3, y=0.6, text="Ambiguous", showarrow=False, font=dict(size=14, color="black"))
fig.add_annotation(x=0.1, y=0.2, text="Hard-to-Learn", showarrow=False, font=dict(size=14, color="black"))

# Show plot
fig.show()
```
